```python
def fetch_market_cap(ticker_list):
    data=yf.download(ticker_list,period='1d',group_by='tickers',auto_adjust=False)


    stock_data=[]
    for tickers in ticker_list:
        data_price=data[tickers]['Close'].iloc[-1]
        market_cap=yf.Ticker(tickers).info.get("marketCap","N/A")
        stock_data.append({
            "Ticker":tickers,
            "Market Cap":market_cap,
            "Latest Price":data_price,
        })
    stocks_df=pd.DataFrame(stock_data)
    return stocks_df


tickers_list=tickers["Ticker"].values.tolist()
df=fetch_market_cap(tickers_list)
df.head()
```

```python
df=df.sort_values(by="Market Cap",ascending=False)
df
```
